# Time Series - Fundamental Concepts

A *time series* is a specific type of data, where measurements of a single quantity are taken over time at regular intervals.

A difficulty associated analyzing time-series is that very often each data point is partially a function of previous datapoints. This breaks one assumption of linear regression: that the targets aren't correlated with each other.

When speaking in equations, we will generally represent time with an index $i$, and the observations from the series as $y_i$.  So the time series as a whole is

$$y_1, y_2, y_3, \ldots $$

You can often use a regression model(\*) to capture a general trend in the series.

As a first pass, let's try to capture a linear trend in the python search data by fitting a simple linear model - with the time *t* as the only independent variable - and use its predictions on the training data.

```python
def to_col_vector(arr):
    """Convert a one dimensional numpy array to a column vector."""
    return arr.reshape(-1, 1)

def make_design_matrix(arr):
    """Constructs our design matrix from a numpy array; includes an intercept term
    and puts it in the expected dimensionality"""
    return sm.add_constant(to_col_vector(arr))
```

```python
def fit_linear_trend(series):
    """Fit a linear trend to a time series.  Return the fit trend as a numpy array."""
    X = make_design_matrix(np.arange(len(series)) + 1)
    y = series.values
    linear_trend_ols = sm.OLS(y, X).fit()
    linear_trend = linear_trend_ols.predict(X)
    return linear_trend
```

```python
# Optional: Showing the OLS setup above with pandas
y = pd.DataFrame(google_trends['python'])
X = make_design_matrix(np.arange(len(y)) + 1)
X = pd.DataFrame(X, columns=['constant','t'], index=y.index)
X.head()
# y.head()
```

```python
def plot_linear_trend(ax, name, series):
    linear_trend = fit_linear_trend(series)
    plot_trend_data(ax, name, series) # From earlier
    ax.plot(series.index, linear_trend)
```

```python
fig, ax = plt.subplots(1, figsize=(14, 2))
plot_linear_trend(ax, 'python', google_trends['python'])
plt.tight_layout()
```

If we subtract out the fitted trend from the original series, we get the **detrended series**.

```python
python_series = google_trends['python']
python_linear_trend = fit_linear_trend(python_series)
python_series_detrended = python_series - python_linear_trend

fig, ax = plt.subplots(1, figsize=(14, 2))
ax.plot(python_series_detrended.index, python_series_detrended)
ax.set_title("Google Searches for Python, Detrended")
plt.tight_layout()
```

Detrending a series is often the first step in analysing a time series.

## Trend Fitting: Moving Average

A very general approach that can be used for detrending data is to compute a **moving average** (also known as a **rolling average**)

The moving average estimate of the trend at a data point $y_i$ is

$$ \hat y_i = \frac{1}{w} \sum_{j = -w/2}^{w/2} y_{i + j} $$

We esentially slide a *window* of a fixed size across our data, and average the values of the series within the window.

The parameter $w$ controls how far to the left and to the right of $i$ we look when averaging the nearby points, this is called the **window**.

![title](img/ra.gif)

```python
# Inputting the toy data above
df = pd.DataFrame({'value': [1,2,3,15,5,6]})
df['moving_average'] = df.rolling(3, center=True).mean()
df
```

**Question**: Why do we have NaNs at the ends of our data?

Let's try a moving average on some google trends.

```python
# Moving to functions, note the default window here is 6
def fit_moving_average_trend(series, window=6):
    return series.rolling(window, center=True).mean()

def plot_moving_average_trend(ax, name, series, window=6):
    moving_average_trend = fit_moving_average_trend(series, window)
    plot_trend_data(ax, name, series)
    ax.plot(series.index, moving_average_trend)
```

```python
fig, axs = plt.subplots(3, figsize=(14, 6))
plot_moving_average_trend(axs[0], 'python', google_trends['python'])
plot_moving_average_trend(axs[1], 'blackberry', google_trends['blackberry'])
plot_moving_average_trend(axs[2], 'gmail', google_trends['gmail'])
plt.tight_layout()
```

Varying the window parameter changes how closely the rolling average fits to the data:

```python
fig, axs = plt.subplots(3, figsize=(14, 12))

gmail_series = google_trends['gmail']
for i, window in enumerate([4, 12, 24]):
    axs[i].plot(gmail_series.index, gmail_series)
    axs[i].plot(gmail_series.index, fit_moving_average_trend(gmail_series, window=window))
    axs[i].set_title("Moving Average Smoothed Gmail Search Data with Window = {} weeks".format(window))
    
plt.tight_layout()
```

We see that:

  - Smaller values of `window` will tend to be influenced by noise of other non-trend patterns in the series.
  - Large values of `window` produce smoother estimates of the general trend in the data.
 
For this reason, when estimating the trend component of time series, we generally prefer larger windows that don't capture the short term variations.

*Aside*: Does this tradeoff remind you of any other tradeoffs we've discussed?

When we have data that aligns with calendar regularities (quarterly, weekly, yearly), it is a good idea to chose the window so that entire cycle is used to find the trend and contained within the window. This will average out any **seasonal** patterns in the data, as we will discuss below.

```python
# 52 weeks in a year
fig, axs = plt.subplots(3, figsize=(14, 6))
plot_moving_average_trend(axs[0], 'python', google_trends['python'], window=52)
plot_moving_average_trend(axs[1], 'blackberry', google_trends['blackberry'], window=52)
plot_moving_average_trend(axs[2], 'gmail', google_trends['gmail'], window=52)
plt.tight_layout()
```

(*Aside*: We can make windows for other aggregate features, such as a rolling *max* or a rolling *variance*, which can be very useful in other analyses)

## Seasonality

A **seasonal** pattern in a time series is one that tends to appear regularly, and often aligns with features of the calendar.

```python
fig, axs = plt.subplots(2, figsize=(14, 4))
plot_trend_data(axs[0], 'baseball', google_trends['baseball'])
plot_trend_data(axs[1], 'taxes', google_trends['taxes'])
plt.tight_layout()
```

Just like we can de**trend** a time series, we can also **deseasonalize** a time series.

The simplest method is to create dummy variables at regular intervals of the calendar, for example:
  - A dummy for each month.
  - A dummy for each season.
  - A dummy for each day of the week (perhaps capturing weekend or weekday trends)
  
and then fit a linear regression to the series using these dummy variables.

```python
baseball_series = google_trends['baseball']
baseball_series
```

```python
# Let's try capturing monthly seasonality
# Creating dummy indicators for the month.
month = baseball_series.index.month
dummies = pd.get_dummies(month, prefix='month').iloc[:,:11]
dummies
```

```python
# Convert to numpy and add the constant term
X = sm.add_constant(dummies.values)
```

```python
y = baseball_series.values
seasonal_model = sm.OLS(y, X).fit()
season = seasonal_model.predict(X)
new_df = pd.DataFrame(data=season, index=baseball_series.index, columns=['date'])
new_df[:10]
# The predicted value only changes when the month changes
```

```python
fig, axs = plt.subplots(2, figsize=(14, 4))
axs[0].plot(baseball_series.index, baseball_series, label='data', c='k')
axs[1].plot(baseball_series.index, season, label='fit season', c='r')
fig.suptitle("Seasonality Decomposition of Baseball Interest")
fig.legend();
```

*Aside*: This dummy variable process is *equivalent* to just grouping the values by month and averaging them: convincing yourself of this is a good exercise for the reader (I've always wanted to say that).

```python
month_means = baseball_series.groupby(baseball_series.index.month).mean()
month_means.rename_axis('month')[9:12]
```

### Put all of this in functions

```python
def create_monthly_dummies(series):
    month = series.index.month
    # Only take 11 of the 12 dummies to avoid strict colinearity.
    return pd.get_dummies(month).iloc[:, :11]
```

```python
def fit_seasonal_trend(series):
    dummies = create_monthly_dummies(series)
    X = sm.add_constant(dummies.values, prepend=False)
    seasonal_model = sm.OLS(series.values, X).fit()
    return seasonal_model.predict(X)
```

```python
def plot_seasonal_trend(ax, name, series):
    seasons_average_trend = fit_seasonal_trend(series)
    plot_trend_data(ax, name, series)
    ax.plot(series.index, seasons_average_trend)
```

Here are linear regression models fit to the baseball and taxes trends with monthly dummy variables.

```python
fig, axs = plt.subplots(2, figsize=(14, 5))

plot_seasonal_trend(axs[0], 'baseball', google_trends['baseball'])
plot_seasonal_trend(axs[1], 'taxes', google_trends['taxes'])
plt.tight_layout()
```

To deseasonalize our series, we simply subtract out our seasonal predictions.

```python
baseball_series = google_trends['baseball']
baseball_seasonal_trend = fit_seasonal_trend(baseball_series)
baseball_seasonal_detrended = baseball_series - baseball_seasonal_trend

fig, ax = plt.subplots(1, figsize=(14, 2))
ax.plot(baseball_series.index, baseball_seasonal_detrended)
ax.set_title("Google Searches for Baseball, Deseasonalized")
```

```python
tax_series = google_trends['taxes']
tax_seasonal_trend = fit_seasonal_trend(tax_series)
tax_seasonal_detrended = tax_series - tax_seasonal_trend

fig, ax = plt.subplots(1, figsize=(14, 2))
ax.plot(tax_series.index, tax_seasonal_detrended)
ax.set_title("Google Searches for Taxes, Deseasonalized")
```

## Trend-Seasonal-Residual Decomposition

The **Classical Trend-Seasonal-Residual Decomposition** expresses a time series as the sum of three components:

$$ y_t = T_t + S_t + R_t $$

Suppose that we are working with *weekly* data, so that each $52$ observations form a calendar year. Then, the initial series $y_t$ is decomposed as follows:

1. Compute the trend component $T_t$ using a moving average with window width $52$ (or $12$ for monthly data).  Then detrend the series ($y_t - T_t$)
2. Compute the seasonal component $S_t$ of the detrended series $y_t - T_t$ by averaging together the observations that fall in the same week/month/year (or making dummy variables and running OLS, but again they're equivalent)
3. The **remainder**, or **error**, or **residual** series $E_t$ is $y_t - T_t - S_t$. This is the "noise".

**Forecasting**: 
<br>
To make a prediction in the future, add the expected trend value and seasonal value at that point in time in the future. (And don't add the residual - it's noise).

Statsmodels implements the classical decomposition as `seasonal_decompose`.

```python
python_decomposition = sm.tsa.seasonal_decompose(google_trends['python'])
```

```python
def plot_seasonal_decomposition(axs, series, sd):
    axs[0].plot(series.index, series)
    axs[0].set_title("Raw Series")
    axs[1].plot(series.index, sd.trend)
    axs[1].set_title("Trend Component $T_t$")
    axs[2].plot(series.index, sd.seasonal)
    axs[2].set_title("Seasonal Component $S_t$")
    axs[3].plot(series.index, sd.resid)
    axs[3].set_title("Residual Component $R_t$")
```

```python
fig, axs = plt.subplots(4, figsize=(14, 8))
plot_seasonal_decomposition(axs, google_trends['python'], python_decomposition)
plt.tight_layout()
```

This decomposition validates our description of the series earlier as a general upwards trend, but with some seasonal behaviour (like a sharp drop off in queries around wintertime).

When the decomposition is working well, the residual component $R_t$ should show no seasonal or trend patterns.

```python
baseball_decomposition = sm.tsa.seasonal_decompose(google_trends['baseball'])

fig, axs = plt.subplots(4, figsize=(14, 8))
plot_seasonal_decomposition(axs, google_trends['baseball'], baseball_decomposition)
plt.tight_layout()
```

**Note:** The different components of the series are often on very different scales, so make sure to pay close attention to the $y$-axis labels.

The classical decomposition is a good first step in peaking at a series, but it's often not powerful enough to capture more complicated trends or seasonal components of a time series.

For example, the gmail series decomposition shows some interesting behaviour in the residuals:

```python
gmail_decomposition = sm.tsa.seasonal_decompose(google_trends['gmail'])

fig, axs = plt.subplots(4, figsize=(14, 8))
plot_seasonal_decomposition(axs, google_trends['gmail'], gmail_decomposition)
plt.tight_layout()
```

Around October 2013 the series showed a clear and sudden change in average level, which the trend component could not adapt to quickly enough.  Consequently, this discontinuity appears in the residual series.

A more sophisticated procedure should be used to decompose this series, such as one that can handle discontinuities like this, or try a more sophisticated model for the trend component.

### WARNING: Train-Holdout Splits with Time Series

A warning, and an example of where time series requires a different approach:

For the supervised learning approaches, you _must_ split along consecutive time frames rather than randomly assigning days to each set if you're forecasting. If you have a year of data, you want to do something like:

*Train*: January 1st - October 31st  
*Test*: November 1st - December 31st

If you simply randomly assign days, features in the training set will be strongly influenced by holdout datapoints (and vice versa) and constitute a huge data leak. You'll get excellent model scores!... but probably terrible scores once you try your model on completely unseen data.

```python
# A year's worth of baseball data
baseball = load_google_trend_data('baseball')['2012-01-01':'2013-01-01']
```

```python
# NO NO NO BAD
from sklearn.model_selection import train_test_split
train_dates, test_dates = train_test_split(baseball)
```

```python
# Yes
train_dates = baseball['2012-01-01':'2012-10-31']
test_dates = baseball['2012-11-01':'2013-01-01']
```
