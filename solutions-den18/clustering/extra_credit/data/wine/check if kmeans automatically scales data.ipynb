```python
# VERDICT: IT DOESN'T SCALE THE DATA FOR YOU
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs
import matplotlib.pyplot as py
%matplotlib inline
```

```python
X, y = make_blobs(n_samples=1000,n_features=2,centers=3)
```

```python
py.scatter(X[:,0][y==0],X[:,1][y==0],c='r')
py.scatter(X[:,0][y==1],X[:,1][y==1],c='b')
py.scatter(X[:,0][y==2],X[:,1][y==2],c='g')
```

```python
model = KMeans(n_clusters=3)
model.fit(X)
py.scatter(X[:,0][model.labels_==0],X[:,1][model.labels_==0],c='r')
py.scatter(X[:,0][model.labels_==1],X[:,1][model.labels_==1],c='b')
py.scatter(X[:,0][model.labels_==2],X[:,1][model.labels_==2],c='g')
```

```python
X[:,0] = X[:,0]*.001
X[:,1] = X[:,1]*1000
model = KMeans(n_clusters=3)
model.fit(X)
py.scatter(X[:,0][model.labels_==0],X[:,1][model.labels_==0],c='r')
py.scatter(X[:,0][model.labels_==1],X[:,1][model.labels_==1],c='b')
py.scatter(X[:,0][model.labels_==2],X[:,1][model.labels_==2],c='g')
```
