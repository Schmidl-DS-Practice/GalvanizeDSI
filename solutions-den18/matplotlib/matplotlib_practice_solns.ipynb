```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
#plt.style.use('ggplot') #uncomment this to change the plotting style
```

### 1. Plot some data

```python
x = np.random.randint(1000, size=50)
y = np.random.randint(1000, size=50)
```

```python
even_sum = (x + y)%2
even_sum
```

```python
fig, ax = plt.subplots()
cax = ax.scatter(x, y, c=even_sum, s=50, alpha=0.7)
ax.axis([-10,1010,-10,1010])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('fake data!')
cbar = fig.colorbar(cax)
```

### 2. Plot some functions

```python
x2 = np.linspace(0, 5, 50)
```

```python
y3 = 3*x2 + 0.5
y4 = 5*np.sqrt(x2)
```

```python
fig, ax = plt.subplots()
ax.plot(x2, y3, 'k--*', label='linear')
ax.plot(x2, y4, 'g-^', label='sqrt')
ax.legend(loc='best')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('some functions')
```

```python
fig, ax = plt.subplots()

ax.plot(x2, y3, 'k--*', label='linear')
ax.plot(x2, y4, 'g-^', label='sqrt')
ax.legend(loc='best')
ax.set_xscale('log')
#ax.set_yscale('log')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('some functions')
```

### 3. Bar plot

```python
fig, ax = plt.subplots()
barheights = [3,5,1]
barlabels = ['grapes', 'oranges', 'hockey pucks']
ax.bar(range(len(barheights)), barheights)
ax.set_xticks(range(len(barheights)))
ax.set_xticklabels(barlabels, rotation=45)
```

### 4. Subplots

```python
fig, axs = plt.subplots(2, 2, figsize=(8,8))

ax = axs[0, 0]
cax = ax.scatter(x, y, c=even_sum, s=50, alpha=0.7)
ax.axis([-10,1010,-10,1010])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('fake data!')
fig.colorbar(cax, ax=ax)

ax = axs[0, 1]
ax.plot(x2, y3, 'k--*', label='linear')
ax.plot(x2, y4, 'g-^', label='sqrt')
ax.legend(loc='best')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('some functions')

ax = axs[1, 0]
ax.plot(x2, y3, 'k--*', label='linear')
ax.plot(x2, y4, 'g-^', label='sqrt')
ax.legend(loc='best')
ax.set_xscale('log')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('some functions')

ax = axs[1, 1]
ax.bar(range(len(barheights)), barheights)
ax.set_title('things in my jeep')
ax.set_xticks(range(len(barheights)))
ax.set_xticklabels(barlabels, rotation=45)

plt.tight_layout()
plt.savefig('great_job.png')
```
